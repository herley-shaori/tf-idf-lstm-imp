*word_set* adalah himpunan yang menampung seluruh kata dari koleksi dokumen (dokumen latih dan dokumen uji).
<br>Terdapat **24256** kata yang tersebar dalam **1118** kalimat</br>

```python
data_array = data.teks.values
data_array = [x.lower() for x in data_array]
word_set = set()
word_split_list = list()
for x in data_array:
    a = x.split(" ")
    a = [k.replace("'",'') for k in a]
    word_split_list.append(a)
    a = set(a)
    word_set = word_set.union(a)

word_set.remove('')
print("Jumlah Kata: ", len(word_set))
print(word_set)
```

Langkah ini menghasilkan matrix sebelum menjadi tf-idf.
<br>Mohon bersabar, langkah ini memakan waktu yang agak lama, sesuai kapasitas komputer yang digunakan.</br>

```python
word_dict_list = list()
for x in data_array:
    word_dict_list.append(dict.fromkeys(word_set, 0))

for x in range (0, len(word_dict_list)):
    wordDict = word_dict_list[x]
    kalimat = word_split_list[x]
    for word in kalimat:
        if(word):
            wordDict[word]+=1
matrix = pd.DataFrame(word_dict_list)
matrix
```

**Memulai Perhitungan TF-IDF**

```python
def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict

def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict 

def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf
```

```python
# TF
tf_list = list()
for x in range(0, len(word_dict_list)):
    tf_list.append(computeTF(word_dict_list[0],word_split_list[0]))
    
# IDF
idfs = computeIDF(word_dict_list)

# TF-IDF
tf_idf_list = list()
for x in tf_list:
    tf_idf_list.append(computeTFIDF(x, idfs))
```

```python
tfidf = pd.DataFrame(tf_idf_list)
```
